```python
import matplotlib.pyplot as plt
import pandas as pd

# 데이터 히스토그램 그리기 위한 함수
def get_plot_distribution(series, plot_name='', xlabel='', ylabel=''):
    plt.figure(figsize=(6, 4))
    
    # 수치형 데이터인 경우
    if pd.api.types.is_numeric_dtype(series):
        series.plot(kind='hist', bins=10, color='blue', edgecolor='black')
        plt.title(plot_name)
        plt.xlabel(xlabel)
        plt.ylabel(ylabel)
    # 카테고리형 데이터인 경우
    else:
        value_counts = series.value_counts()
        value_counts.plot(kind='bar', color=['blue', 'red'])
        plt.title(plot_name)
        plt.xlabel(xlabel)
        plt.ylabel(ylabel)
        plt.xticks(rotation=0)
    
    plt.show()

# 데이터 간의 관계 파악하기 위한 함수
def get_corr(series1, series2, min_count=10, bins=10, plot_name='', xlabel='', ylabel=''):
    # 각 범주의 데이터 개수 계산
    value_counts = series1.value_counts()
    
    # 데이터가 너무 적은 데이터 제거 -> 두 series 간의 관계 파악이 잘못 될 가능성이 높기 때문
    filtered_series1 = series1[series1.isin(value_counts[value_counts > min_count].index)]
    filtered_series2 = series2[series1.isin(value_counts[value_counts > min_count].index)]
    
    # 수치형 데이터인 경우
    if pd.api.types.is_numeric_dtype(series1):
        series_bins = pd.cut(filtered_series1, bins=bins)
        crosstab = pd.crosstab(series_bins, filtered_series2, normalize='index')
    # 범주형 데이터일 경우
    else:
        crosstab = pd.crosstab(filtered_series1, filtered_series2, normalize='index')
    
    crosstab.plot(kind='bar', stacked=True, color=['skyblue', 'lightcoral'], figsize=(8, 6))
    
    plt.title(plot_name)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)

    unique_values = series2.unique()
    legend_labels = [f'{val}' for val in unique_values]
    plt.legend(legend_labels)
    plt.grid(True)
    plt.show()
```

```python
import scipy.stats

thre = 0.05
significant_results = []
for i in range(len(get_dum_df.columns)):
    for j in range(i+1, len(get_dum_df.columns)):
        # i 번째 컬럼이 1인 경우의 뇌졸중 데이터를 추출
        col_values = get_dum_df['stroke'][get_dum_df.iloc[:, i] == 1].values
        
        # 'work_type_Never_worked'가 1인 경우의 뇌졸중 데이터를 추출
        row_values = get_dum_df['stroke'][get_dum_df.iloc[:, j] == 1].values
        
        # 두 그룹 간의 t-테스트 수행
        t_stat, p_val = scipy.stats.ttest_ind(col_values, row_values)
        
        if p_val < thre:
            significant_results.append((get_dum_df.iloc[:, i].name, get_dum_df.iloc[:, j].name, t_stat, p_val)) 
print(len(significant_results))
for result in significant_results:
    print(f"{result[0]} Compare {result[1]}, t-statistic: {result[2]}, p-value: {result[3]}")     
```
